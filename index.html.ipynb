```python
<!DOCTYPE html>
<html>
<head>
    <title>Diagnosis Keperawatan</title>
    <style>
        body {
            font-family: Arial;
            background-color: #eef6f9;
            padding: 20px;
        }
        .card {
            background: white;
            padding: 20px;
            border-radius: 10px;
            width: 350px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }
        h1 {
            color: #2c3e50;
        }
        button {
            background-color: #27ae60;
            color: white;
            padding: 10px;
            border: none;
            border-radius: 5px;
            margin-top: 10px;
        }
        button:hover {
            background-color: #1e8449;
        }
        #hasil {
            margin-top: 15px;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <h1>🩺 Sistem Diagnosis Keperawatan</h1>

    <div class="card">
        <h3>Keluhan:</h3>
        <input type="checkbox" id="sesak"> Sesak Napas<br>
        <input type="checkbox" id="nyeri"> Nyeri<br>
        <input type="checkbox" id="cemas"> Cemas<br><br>

        <h3>Tanda Vital:</h3>
        RR: <input type="number" id="rr"><br><br>
        SpO2: <input type="number" id="spo2"><br><br>

        <button onclick="tentukanDiagnosa()">Tentukan Diagnosis</button>

        <div id="hasil"></div>
    </div>

    <script>
        function tentukanDiagnosa() {
            let sesak = document.getElementById("sesak").checked;
            let nyeri = document.getElementById("nyeri").checked;
            let cemas = document.getElementById("cemas").checked;

            let rr = document.getElementById("rr").value;
            let spo2 = document.getElementById("spo2").value;

            let hasil = "";

            if (sesak && rr > 24 && spo2 < 95) {
                hasil = `
                Diagnosis: Gangguan Pertukaran Gas<br>
                Alasan:
                <ul>
                    <li>Sesak napas</li>
                    <li>RR meningkat</li>
                    <li>SpO2 menurun</li>
                </ul>
                Intervensi:
                <ul>
                    <li>Monitor respirasi</li>
                    <li>Berikan oksigen</li>
                    <li>Posisi semi fowler</li>
                </ul>
                `;
            } 
            else if (nyeri) {
                hasil = `
                Diagnosis: Nyeri Akut<br>
                Intervensi:
                <ul>
                    <li>Kaji nyeri</li>
                    <li>Berikan analgesik</li>
                </ul>
                `;
            } 
            else if (cemas) {
                hasil = `
                Diagnosis: Ansietas<br>
                Intervensi:
                <ul>
                    <li>Berikan dukungan emosional</li>
                    <li>Ajarkan relaksasi</li>
                </ul>
                `;
            } 
            else {
                hasil = "Diagnosis tidak ditemukan";
            }

            document.getElementById("hasil").innerHTML = hasil;
        }
    </script>

</body>
</html>
```
